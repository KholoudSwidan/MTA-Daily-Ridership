```python
# تحويل عمود التاريخ إلى نوع datetime
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# تحويل عمود الركاب إلى عدد صحيح
df['Subways: Total Estimated Ridership'] = df['Subways: Total Estimated Ridership'].astype(int)
```

```python
df = df.dropna()
```

```python
# استبدال القيم المفقودة بمتوسط العمود
df.loc[:, 'Subways: Total Estimated Ridership'] = df['Subways: Total Estimated Ridership'].fillna(df['Subways: Total Estimated Ridership'].mean())
```

```python
df.describe()
```

```python
df = df[df['Subways: Total Estimated Ridership'] >= 0]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df['Subways: Total Estimated Ridership'], bins=50, color='blue', alpha=0.7)
plt.xlabel('Number of Riders')
plt.ylabel('Frequency')
plt.title('Distribution of MTA Ridership')
plt.show()
```

```python
#analysis questions we can answer from data 

#how has the ridership changed overtime?


plt.figure(figsize=(12,6))  
plt.plot(df['Date'], df['Subways: Total Estimated Ridership'], label="Subway Ridership", color='pink')
plt.xlabel("Date")
plt.ylabel("Total Ridership")
plt.title("Subway Ridership Trends Over Time")
plt.legend()
plt.grid()
plt.show()
```

```python
#Are weekends and weekdays different in ridership?

# Create a 'Day of Week' column
df['Day_of_Week'] = df['Date'].dt.day_name()

# Group by day of the week
avg_ridership = df.groupby('Day_of_Week')['Subways: Total Estimated Ridership'].mean()

# Reorder days
ordered_days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
avg_ridership = avg_ridership.reindex(ordered_days)

# Plot
plt.figure(figsize=(10,5))
avg_ridership.plot(kind="bar", color="purple")
plt.title("Average Subway Ridership by Day of the Week")
plt.ylabel("Average Ridership")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
 #Has ridership returned to pre-pandemic levels?
    
plt.figure(figsize=(12,6))
plt.plot(df['Date'], df['Subways: % of Comparable Pre-Pandemic Day'], label="Subway Recovery %", color='grey')
plt.plot(df['Date'], df['Buses: % of Comparable Pre-Pandemic Day'], label="Bus Recovery %", color='orange')
plt.xlabel("Date")
plt.ylabel("Recovery Percentage")
plt.title("Public Transport Ridership Recovery (%) Over Time")
plt.legend()
plt.grid()
plt.show()
```

```python
#Which mode of transportation has the highest ridership?


# Calculate the average ridership for each mode
avg_ridership_modes = {
    "Subway": df['Subways: Total Estimated Ridership'].mean(),
    "Buses": df['Buses: Total Estimated Ridership'].mean(),
    "LIRR": df['LIRR: Total Estimated Ridership'].mean(),
    "Metro-North": df['Metro-North: Total Estimated Ridership'].mean(),
    "Staten Island Railway": df['Staten Island Railway: Total Estimated Ridership'].mean(),
}

# Convert to DataFrame
avg_ridership_modes = pd.DataFrame(list(avg_ridership_modes.items()), columns=["Transport Mode", "Avg Ridership"])

# Plot
plt.figure(figsize=(10,5))
plt.bar(avg_ridership_modes["Transport Mode"], avg_ridership_modes["Avg Ridership"], color=["pink", "grey", "orange", "red", "purple"])
plt.xlabel("Transport Mode")
plt.ylabel("Average Daily Ridership")
plt.title("Comparison of Average Daily Ridership by Mode")
plt.grid(axis='y')
plt.show()
```

```python
# Does an increase in car traffic reduce public transit ridership?

import seaborn as sns

# Compute correlation matrix
correlation_matrix = df[['Subways: Total Estimated Ridership', 'Buses: Total Estimated Ridership', 
                         'LIRR: Total Estimated Ridership', 'Metro-North: Total Estimated Ridership',
                         'Bridges and Tunnels: Total Traffic']].corr()

# Plot heatmap
plt.figure(figsize=(8,5))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Between Car Traffic and Public Transport Ridership")
plt.show()
```

```python
#What are the busiest months for ridership?

# Extract month and year
df['Month'] = df['Date'].dt.month_name()
df['Year'] = df['Date'].dt.year

# Group by Month and calculate average ridership
monthly_ridership = df.groupby('Month')['Subways: Total Estimated Ridership'].mean()

# Reorder months
months_order = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
monthly_ridership = monthly_ridership.reindex(months_order)

# Plot
plt.figure(figsize=(12,6))
plt.plot(monthly_ridership.index, monthly_ridership.values, marker='o', linestyle='-', color='blue')
plt.xlabel("Month")
plt.ylabel("Average Daily Ridership")
plt.title("Average Subway Ridership Per Month")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
#How does ridership vary by time of year? (Seasonal Trend)

# Define seasons
def get_season(month):
    if month in ['December', 'January', 'February']:
        return 'Winter'
    elif month in ['March', 'April', 'May']:
        return 'Spring'
    elif month in ['June', 'July', 'August']:
        return 'Summer'
    else:
        return 'Fall'

df['Season'] = df['Month'].apply(get_season)

# Group by Season
seasonal_ridership = df.groupby('Season')['Subways: Total Estimated Ridership'].mean()

# Sort for better visualization
seasonal_ridership = seasonal_ridership[['Winter', 'Spring', 'Summer', 'Fall']]

# Plot
plt.figure(figsize=(8,5))
seasonal_ridership.plot(kind="bar", color=["blue", "green", "orange", "red"])
plt.title("Subway Ridership by Season")
plt.ylabel("Average Daily Ridership")
plt.grid(axis='y')
plt.show()
```

```python
#How do different transit modes compare daily?

plt.figure(figsize=(12,6))
plt.plot(df['Date'], df['Subways: Total Estimated Ridership'], label="Subway", color='blue')
plt.plot(df['Date'], df['Buses: Total Estimated Ridership'], label="Bus", color='green')
plt.plot(df['Date'], df['LIRR: Total Estimated Ridership'], label="LIRR", color='red')
plt.plot(df['Date'], df['Metro-North: Total Estimated Ridership'], label="Metro-North", color='orange')

plt.xlabel("Date")
plt.ylabel("Daily Ridership")
plt.title("Ridership Trends for Different Transit Modes")
plt.legend()
plt.grid()
plt.show()
```

```python
#Impact of COVID on Public Transport Usage

plt.figure(figsize=(12,6))
plt.plot(df['Date'], df['Subways: % of Comparable Pre-Pandemic Day'], label="Subway Recovery", color='blue')
plt.plot(df['Date'], df['Buses: % of Comparable Pre-Pandemic Day'], label="Bus Recovery", color='green')

plt.xlabel("Date")
plt.ylabel("Percentage of Pre-Pandemic Ridership")
plt.title("Public Transport Recovery Over Time")
plt.legend()
plt.grid()
plt.show()
```
